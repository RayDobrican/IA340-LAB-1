```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the dataframe is named df
df = df.sort_values(['name', 'year'])

# Calculate the year over year income growth rate
df['income_growth_rate'] = df.groupby('name')['income'].pct_change()

# Pivot the data for easier plotting
pivot_df = df.pivot(index='year', columns='name', values='income_growth_rate')

# Plot the data
plt.figure(figsize=(15, 10))
for column in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[column], marker='', linewidth=1, alpha=0.9, label=column)

plt.legend(loc=2, ncol=2)
plt.title("Year over year income growth rate for each state", loc='left', fontsize=12, fontweight=0, color='orange')
plt.xlabel("Year")
plt.ylabel("Income Growth Rate")
plt.show()

#This code first sorts the data by state name and year. It then calculates the year-over-year income growth rate for each state. This is done by grouping the data by state name and using the `pct_change` function to calculate the percentage change in income from the previous year. The code then pivots the data to make it easier to plot. The pivot table has years as the index and state names as the columns, with the income growth rate as the values. Finally, it plots the data on a line chart.
```
